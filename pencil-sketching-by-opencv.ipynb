```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path1 = '/kaggle/input/pencil-sketch-dataset/1.jpg'
path2 = '/kaggle/input/pencil-sketch-dataset/2.jpg'
path3 = '/kaggle/input/pencil-sketch-dataset/3.jpg'
path4 = '/kaggle/input/pencil-sketch-dataset/4.jpg'
path5 = '/kaggle/input/pencil-sketch-dataset/5.jpg'
```

```python
def pencil_sketch(path, k):
    img = cv2.imread(path)
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)
    inv_gray_img = 255 - gray_img
    blur_img = cv2.GaussianBlur(inv_gray_img, (k, k), 0)
    inv_blur_img = 255 - blur_img
    sketch_img = cv2.divide(gray_img, inv_blur_img, scale=256.0)
    
    return rgb_img, sketch_img
```

```python
rgb_img1, sketch_img1 = pencil_sketch(path1, 31)
rgb_img2, sketch_img2 = pencil_sketch(path2, 31)
rgb_img3, sketch_img3 = pencil_sketch(path3, 31)
rgb_img4, sketch_img4 = pencil_sketch(path4, 31)
rgb_img5, sketch_img5 = pencil_sketch(path5, 31)
```

```python
plt.figure(figsize=(7, 5))

plt.subplot(121)
plt.imshow(rgb_img1)
plt.axis('off')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(sketch_img1, cmap='gray')
plt.axis('off')
plt.title('Pencil Sketch')

plt.tight_layout()
plt.show()
```

```python
k_values = [11, 21, 31, 41, 51, 61]

plt.figure(figsize=(12, 8))

for i, k in enumerate(k_values):
    _, sketch_img = pencil_sketch(path1, k)    
    
    plt.subplot(1, len(k_values), i + 1)
    plt.imshow(sketch_img, cmap='gray')
    plt.axis('off')
    plt.title(f'k={k}')

plt.tight_layout()
plt.show()
```
