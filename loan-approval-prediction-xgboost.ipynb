```python
import time
import skimpy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/kaggle/input/playground-series-s4e10/train.csv')
test = pd.read_csv('/kaggle/input/playground-series-s4e10/test.csv')
```

```python
skimpy.skim(train)
```

```python
skimpy.skim(test)
```

```python
train.info()
```

```python
train.drop_duplicates(inplace=True)

train['person_income'] = train['person_income'].astype(float)
train['loan_amnt'] = train['loan_amnt'].astype(float)
```

```python
test.info()
```

```python
test.drop_duplicates(inplace=True)

test['person_income'] = test['person_income'].astype(float)
test['loan_amnt'] = test['loan_amnt'].astype(float)
```

```python
train.info()
```

```python
test.info()
```

```python
train.head()
```

```python
test.head()
```

```python
X = train.drop(['id', 'loan_status'], axis=1)
X_test = test.drop(['id'], axis=1)
y = train['loan_status']
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}

for column in ['person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file']:
    label_encoders[column] = LabelEncoder()
    X[column] = label_encoders[column].fit_transform(X[column])
    X_test[column] = label_encoders[column].transform(X_test[column])
```

```python
X.head()
```

```python
X_test.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    learning_rate=0.1,
    n_estimators=500,
    max_depth=5,
    use_label_encoder=False,
    n_jobs=-1,
    device='cpu'
)
```

```python
start_time = time.time()

model.fit(X_train, y_train)

end_time = time.time()

training_time = end_time - start_time

print("Training time: {:.2f}s".format(training_time))
```

```python
from sklearn.metrics import f1_score, roc_auc_score

y_prob = model.predict_proba(X_val)[:, 1]
y_pred = model.predict(X_val)

f1 = f1_score(y_val, y_pred)
roc_auc = roc_auc_score(y_val, y_prob)

print('roc-auc score is', roc_auc)
print('f1 score is', f1)
```

```python
import seaborn as sb
from sklearn.metrics import confusion_matrix

cfm = confusion_matrix(y_val, y_pred)

sb.heatmap(cfm / np.sum(cfm, axis=1), annot=True, fmt=".2", cmap="Blues")
plt.show()
```

```python
test_pred = model.predict_proba(X_test)[:, 1]

submission = pd.DataFrame({'id': test['id'], 'loan_status': test_pred})
submission.to_csv('submission.csv', index=False)
```
